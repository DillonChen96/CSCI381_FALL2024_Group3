17. <span style="font-size: 11pt; font-family: &quot;Roboto Mono&quot;, monospace; color: rgb(24, 128, 56); background-color: transparent; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; vertical-align: baseline; text-wrap: wrap;">WHERE ct.TotalAmount &gt; 10000</span><span style="font-size: 11pt; background-color: transparent; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; vertical-align: baseline; text-wrap: wrap;">:</span>

```sql
WITH CustomerTotals AS (
    SELECT o.CustomerID, SUM(ol.Quantity * ol.UnitPrice) AS TotalAmount
    FROM Sales.Orders o
    INNER JOIN Sales.OrderLines ol ON o.OrderID = ol.OrderID
    GROUP BY o.CustomerID
)
SELECT c.CustomerName
FROM Sales.Customers c
INNER JOIN CustomerTotals ct ON c.CustomerID = ct.CustomerID
WHERE ct.TotalAmount > 10000;
```

11. <span style="font-size: 11pt; font-family: &quot;Roboto Mono&quot;, monospace; color: rgb(24, 128, 56); background-color: transparent; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; vertical-align: baseline; text-wrap: wrap;">GROUP BY c.CustomerID, c.CustomerName</span><span style="font-size: 11pt; background-color: transparent; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; vertical-align: baseline; text-wrap: wrap;">:</span>

13. <span style="font-size: 11pt; font-family: &quot;Roboto Mono&quot;, monospace; color: rgb(24, 128, 56); background-color: transparent; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; vertical-align: baseline; text-wrap: wrap;">ORDER BY OrderCount DESC</span><span style="font-size: 11pt; background-color: transparent; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; vertical-align: baseline; text-wrap: wrap;">:</span>
    
- <span style="font-size: 11pt; background-color: transparent; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; vertical-align: baseline; text-wrap: wrap;">Sorts the customers in descending order based on </span> <span style="font-size: 11pt; font-family: &quot;Roboto Mono&quot;, monospace; color: rgb(24, 128, 56); background-color: transparent; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; vertical-align: baseline; text-wrap: wrap;">OrderCount</span> <span style="font-size: 11pt; background-color: transparent; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; vertical-align: baseline; text-wrap: wrap;"> so that customers with the most orders appear first.</span>

```sql
SELECT TOP 5 CustomerName, OrderCount
FROM (
    SELECT c.CustomerID, c.CustomerName, COUNT(o.OrderID) AS OrderCount
    FROM Sales.Customers c
    LEFT JOIN Sales.Orders o ON c.CustomerID = o.CustomerID
    GROUP BY c.CustomerID, c.CustomerName
) AS CustomerOrders
ORDER BY OrderCount DESC;


```

```sql
SELECT CustomerName
FROM Sales.Customers
WHERE CustomerID IN (
    SELECT CustomerID
    FROM Sales.Orders
    GROUP BY CustomerID
    HAVING COUNT(OrderID) > 1
);
```
